```python
dogs_arch_clean=dogs_arch_df.copy()
image_pred_clean=image_pred_df.copy()
tweet_json_clean=tweet_json_df.copy()
```

```python
##Cleaning
#quality issues
#1-drop duplicated jpg_url in image prediction
image_pred_clean.drop_duplicates(subset='jpg_url',keep='first',inplace=True)
#test
image_pred_clean.jpg_url.duplicated().sum()
```

```python
###Cleaning Test
image_pred_clean.shape
```

```python
##Cleaning

#1-melt columns doggo, floofer, pupper,puppo into one column called dog_stage column
#code
# handle none
dogs_arch_clean.doggo.replace('None', '', inplace=True)
dogs_arch_clean.floofer.replace('None', '', inplace=True)
dogs_arch_clean.pupper.replace('None', '', inplace=True)
dogs_arch_clean.puppo.replace('None', '', inplace=True)

# merge into column
dogs_arch_clean['dog_stage'] = dogs_arch_clean.doggo + dogs_arch_clean.floofer +dogs_arch_clean.pupper + dogs_arch_clean.puppo

# handle multiple stages
dogs_arch_clean.loc[dogs_arch_clean.dog_stage == 'doggopupper', 'dog_stage'] = 'doggo, pupper'
dogs_arch_clean.loc[dogs_arch_clean.dog_stage == 'doggopuppo', 'dog_stage'] = 'doggo, puppo'
dogs_arch_clean.loc[dogs_arch_clean.dog_stage == 'doggofloofer', 'dog_stage'] = 'doggo, floofer'

# handle missing values
dogs_arch_clean.loc[dogs_arch_clean.dog_stage == '', 'dog_stage'] = np.nan
dogs_arch_clean.drop(['doggo','floofer','pupper','puppo'],inplace=True,axis=1)
```

```python
##Cleaning Test
dogs_arch_clean.shape
dogs_arch_clean.dog_stage.value_counts()
```

```python
##Cleaning
#Tidiness
##2-merge tweet-json with the dogs archive and image prediction data frames
master_clean=pd.merge(dogs_arch_clean,tweet_json_clean,on='tweet_id')
master_clean=pd.merge(master_clean,image_pred_clean,on='tweet_id')
```

```python
###Cleaning test
master_clean.shape
```

```python
#3-Keep original ratings (no retweets) that have images(Twitter Archive Table)
```

```python
##Cleaning
##Code :based on the retweetstatuse id there are 181 retweet iam going to delete the reqcords for retweet
master_clean=master_clean[master_clean['retweeted_status_id'].isnull()]
#test
master_clean.retweeted_status_id.notnull().sum()
master_clean.jpg_url.isnull().sum()
```

```python
##Cleaning Test
master_clean.tweet_id.duplicated().sum()
```

```python
##Cleaning
#4-change time stamp col to date time not string 
#code
master_clean.timestamp=dogs_arch_clean.timestamp.astype('datetime64')
#test
master_clean.info()
```

```python
##Cleaning
#5-source should be a clear text not link like "iphone or android"
#code
master_clean.source=master_clean.source.str.split(r'<|>').str[2]
```

```python
#test
master_clean.source.value_counts()
```

```python
##Cleaning
#6-replace missing values from none to nan in dog stages
#code
master_clean.dog_stage=master_clean.dog_stage.replace('None',np.nan)
#test
master_clean.dog_stage.value_counts()
master_clean.dog_stage.isnull().sum()
```

```python
##Cleaning
#7-Delete columns that wont be used for analysis ('in_reply_to_status_id',
                                                    #'in_reply_to_user_id',text
                                                    #'retweeted_status_id',
                                                    #'retweeted_status_user_id', 
                                                    #'retweeted_status_timestamp', 
                                                    #'expanded_urls')
```

```python
##Cleaning
#code
master_clean.drop(['in_reply_to_status_id','in_reply_to_user_id','retweeted_status_id'
                   ,'retweeted_status_user_id','retweeted_status_timestamp','expanded_urls'],axis=1,inplace=True)
#test
master_clean.head()
```

```python
##Cleaning Test
master_clean.info()
```

```python
##Cleaning
##8-Correct numerators &denominators to be float
#code
master_clean.rating_numerator=master_clean.rating_numerator.astype(float)
master_clean.rating_denominator=master_clean.rating_denominator.astype(float)
#test
master_clean.rating_numerator.dtype
master_clean.rating_denominator.dtype
```

```python
##Cleaning:8 Extract ratings correctly from text

#Code
rating = master_clean.text.str.extract('((?:\d+\.)?\d+)\/(\d+)', expand=True)
rating.columns = ['rating_numerator', 'rating_denominator']
```

```python
##Cleaning Test
master_clean.rating_denominator.value_counts()
master_clean.rating_numerator.value_counts()
```

```python
#Cleaning
#9-retweet_count and favorite_count should be integers, not floats.tweet_id, to object 
master_clean.tweet_id=master_clean.tweet_id.astype('object')
```

```python
#cleaning 
#-Dog Names there are in accurate
#get list of wrong names
names_list=master_clean[master_clean['name'].str.islower() == True].name
#replace wrong names with nan
master_clean['name']=master_clean['name'].replace(to_replace=names_list,value=np.nan)
#test
master_clean[master_clean['name'].str.islower() == True]
```
