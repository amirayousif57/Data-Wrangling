```python
##Cleaning
```

```python
dogs_arch_clean=dogs_arch_df.copy()
image_pred_clean=image_pred_df.copy()
tweet_json_clean=tweet_json_df.copy()
```

```python
#quality issues
#1-drop duplicated jpg_url in image prediction
image_pred_clean.drop_duplicates(subset='jpg_url',keep='first',inplace=True)
#test
image_pred_clean.jpg_url.duplicated().sum()
```

```python
image_pred_clean.shape
```

```python
#2-rating dominator should be 10
#code replace all values above or below 10 to be 10 

#df1['A'].values[df1['A'] > 9] = 11
dogs_arch_clean['rating_denominator'].values[dogs_arch_clean['rating_denominator']>10]=10
dogs_arch_clean['rating_denominator'].values[dogs_arch_clean['rating_denominator']<10]=10
#test
dogs_arch_clean.query('rating_denominator>10')
```

```python
# iwill solve tidiness issues first to avoid double work
#Tidiness
##1-melt columns doggo, floofer, pupper,puppo into one column called dog_stage column
#['doggo','floofer','pupper','puppo']
dogs_arch_clean=pd.melt(dogs_arch_clean,id_vars=['tweet_id',
                                                 'in_reply_to_status_id','in_reply_to_user_id','timestamp','source','text','retweeted_status_id',
                                                 'retweeted_status_user_id','retweeted_status_timestamp','expanded_urls',
                                                 'rating_numerator','rating_denominator','name'],
                        var_name='vardogs',value_name='dog_stage')
#then drop the vardogs
dogs_arch_clean.drop('vardogs',axis=1,inplace=True)
dogs_arch_clean = dogs_arch_clean.sort_values('dog_stage').drop_duplicates(subset='tweet_id', 
                                                                                        keep='last')
```

```python
dogs_arch_clean.shape
dogs_arch_clean.dog_stage.value_counts()
#it seems there are morve stages than the 4 stages and it wont be
```

```python
#Tidiness
##2-merge tweet-json with the dogs archive and image prediction data frames
master_clean=pd.merge(dogs_arch_clean,tweet_json_clean,on='tweet_id')
master_clean=pd.merge(master_clean,image_pred_clean,on='tweet_id')
```

```python
#test
master_clean.shape
```

```python
master_clean.shape
```

```python
#3-Keep original ratings (no retweets) that have images(Twitter Archive Table)
```

```python
##Code :based on the retweetstatuse id there are 181 retweet iam going to delete the reqcords for retweet
master_clean=master_clean[master_clean['retweeted_status_id'].isnull()]
#test
master_clean.retweeted_status_id.notnull().sum()
master_clean.jpg_url.isnull().sum()
```

```python
master_clean.tweet_id.duplicated().sum()
```

```python
#4-change time stamp col to date time not string 
#code
master_clean.timestamp=dogs_arch_clean.timestamp.astype('datetime64')
#test
master_clean.info()
```

```python
#5-source should be a clear text not link like "iphone or android"
#code
master_clean.source=master_clean.source.str.split(r'<|>').str[2]
```

```python
#test
master_clean.source.value_counts()
```

```python
#6-replace missing values from none to nan in dog stages
#code
master_clean.dog_stage=master_clean.dog_stage.replace('None',np.nan)
#test
master_clean.dog_stage.value_counts()
master_clean.dog_stage.isnull().sum()
```

```python
#7-Delete columns that wont be used for analysis ('in_reply_to_status_id',
                                                    #'in_reply_to_user_id',text
                                                    #'retweeted_status_id',
                                                    #'retweeted_status_user_id', 
                                                    #'retweeted_status_timestamp', 
                                                    #'expanded_urls')
```

```python
#code
master_clean.drop(['in_reply_to_status_id','in_reply_to_user_id','retweeted_status_id'
                   ,'retweeted_status_user_id','retweeted_status_timestamp','expanded_urls'],axis=1,inplace=True)
#test
master_clean.head()
```

```python
master_clean.info()
```

```python
##8-Correct numerators to be float
#code
master_clean.rating_numerator=master_clean.rating_numerator.astype(float)
#test
master_clean.rating_numerator.dtype
```

```python
master_clean.rating_denominator.isnull().sum()
```

```python
master_clean.rating_denominator.value_counts()
```

```python
master_archive=pd.read_csv("twitter_archive_master.csv")
```

```python
#1- Get the rating mean for top5 dog stages by rating means
top5=master_archive.groupby('dog_stage')['rating_numerator'].mean()
top5=top5.sort_values()
```

```python
print(top5)
```

```python
#2- get the  popular tweets of dog stages this can be known by retweet count
populairty=master_archive.groupby('dog_stage')['retweet_count'].count()
populairty
```

```python
#3-used sources to rate dogs
popular_sources=master_archive.source.value_counts()
print(popular_sources)
```

```python
#4- image predictions vs favourits count
img_fav=master_archive.groupby('p1')['favorite_count'].count()
```

```python
img_fav=img_fav.sort_values(ascending=False).head(5)
print(img_fav)
#imgvscount=pd.DataFrame(img_fav)
#imgvscount
```

```python
#top rated tweets by favourite count
top1=master_archive.sort_values(by='favorite_count').head(1)
print("most loved tweet :{} {}".format(top1['text'],top1['jpg_url']))
least1=master_archive.sort_values(by='favorite_count').tail(1)
print("least Loved tweet :{} {}".format(least1['text'],least1['jpg_url']))
```
